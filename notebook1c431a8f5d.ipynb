# Dealing with na 

```python
#find the mode of iso2 column
mode=df['iso2'].mode().sort_values()[0]
```

```python
#fill the na value in iso2
df['iso2'].fillna('mode')
```

```python
#stripping the object columns

df['iso2'].str.strip()
df['region'].str.strip()
df['capital'].str.strip()
df['currency'].str.strip()
```

```python
#drop rows with more than 10 na
df=df.dropna(thresh=10,axis=0)
```

```python
#fill na in numerical columns with their median
numerical_cols=df.select_dtypes(include=['float64']).columns
for col in numerical_cols:
    group_median=df.groupby('region')[col].transform('median')
    global_median=df[col].median()
    df[col]=df[col].fillna( group_median).fillna(global_median)
```

# gdp vs infant mortality
###  Objective: To determine the efficacy of macroeconomic wealth in eradicating premature childhood deaths and earlymedical vulnerabilities.


```python
#plot the scatter plot
gdp_val=np.log(df['gdp'])
im=df['infant_mortality'].apply(lambda x:np.log(x))
plt.scatter(gdp_val,im,color='blue')
plt.xlabel('gdp')
plt.ylabel('infant mortality')
plt.title('HOW GDP CORRELATES WITH INFANT MORTALITY')
plt.grid(True,which='both',ls='--',alpha=0.5)
plt.show()
```

### Negative Correlation. Infant mortality drops  steeply with early increases in GDP. Basic economic expansion allows funding for clean water, sanitation, and infant vaccinations, saving lives.

industrilazation and environmental hazards

### Objective: To identify if economic growth leads to an escalation of greenhouse gas outputs or if sustainable decoupling is occurring.

```python
#plot the scatter plot for gdp vs co2 emmision
gdp_val=np.log(df['gdp'])
co2=df['co2_emissions'].apply(lambda x:np.log(x))
plt.scatter(gdp_val,co2,color='red')
plt.xlabel('gdp')
plt.ylabel('co2_emissions')
plt.title('HOW GDP CORRELATES WITH CO2 EMISSION')
plt.grid(True,which='both',ls='--',alpha=0.5)
plt.show()
```

### Linear Positive Correlation. Higher GDP strongly pairs with higher emissions. Industrial output relies heavily on carbon energy inputs.
